```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('code_final/RF/madelon2D.csv')
df100 =pd.read_hdf("code_final/PCA/datasets.hdf", 'madelon')
```

```python
# df.target_original.replace({'RB':0 , 'NRB': 1}, inplace=True)
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.Number, df.ICA, label='+NN')
plt.plot(df.Number, df.PCA, label='PCA+NN')
plt.plot(df.Number, df.RF, label='RF+NN')
plt.plot(df.Number, df.RP, label='RP+NN')

plt.legend()

plt.xlabel('# of Clusters')
plt.ylabel('Fit Time')
plt.title('Fit Time v/s # of Clusters Cancer')
plt.show()
```

```python
plt.scatter(df.x, df.y, c=df.target, alpha=0.8)
plt.xlabel("")
plt.ylabel("")
plt.title("Actual Clusters after RF")
plt.show()
```
